A continuación se proponen una serie de ejercicios con funciones para que practiques. Pero recordemos primero cómo se define una función y cómo se utiliza. En el siguiente ejemplo, 'cuadrado' es el nombre de mi función, 'x' es el argumento y como hay un 'return' sabremos que devuelve un valor a la salida cuando usamos dicha función.

```python
def cuadrado(x):
  return x**2

y = 4
y2 = cuadrado(y)
print('El cuadrado de', y, 'es', y2)
```

**None** es un tipo de datos especial que significa valor nulo o vacío. Es muy útil en el preprocesado de datos o cuando queremos devolver en una función un error.

```python
cosa = None
if cosa is None:
    print("Es nada")
else:
    print("Es algo")
```

```python
#<SOL>
def orden_inverso(numero):
  numero_str = str(numero)
  for i in range(len(numero_str)-1, -1, -1):
    print(numero_str[i])


numero = 12345
print('el número', numero, 'al revés es')
orden_inverso(numero)

#</SOL>
```

> **Ejercicio EXTRA**: Diseñe una función que nos diga si un número dado como parámetro es perfecto o no. Un número es perfecto si es igual a la suma de todos sus divisores excepto él mismo.




```python
# <SOL>
def es_perfecto(num):
  sumatorio = 0
  for i in range(1, num):
    if num % i == 0:
      sumatorio += i
  
  return sumatorio == num

numero = 29
print('Número', numero, 'perfecto?:', es_perfecto(numero))
# </SOL>
```

### 1.4 Especificar parámetros con valores por defecto (Extra!)

Es posible especificar valores por defecto en los parámetros de una función. El valor por defecto se usará cuando en la llamada a la función no se haya proporcionado el correspondiente argumento.

```python
def menu(bebida, entrante, postre='Tiramisú'):
    return {'bebida': bebida, 'entrante': entrante, 'postre': postre}
```

```python
# Hacemos uso del valor por defecto del parámetro "postre"
menu('Vino', 'Ensalada')
```

```python
# "Sobreescribimos" el valor de "postre" especificando uno concreto
menu('Agua', 'Revuelto de setas', 'Helado')
```

Los valores por defecto en los parámetros se calculan cuando se **define** la función, no cuando se **ejecuta**.
